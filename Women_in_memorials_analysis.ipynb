```python
import pandas as pd
import pymorphy3
import re
morph = pymorphy3.MorphAnalyzer()
```

```python
df = pd.read_excel('data-2801-2024-03-20.xlsx').drop([0])
df
```

```python
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

# Remove punctuation from the "Name" column
df['Name'] = df['Name'].apply(remove_punctuation)
df
```

```python
def check_for_single_name_in_line(line):
    line = line.split()
    for i in range(len(line)-2):
        # print(line[i], morph.parse(line[i])[0].tag)
        if (("Surn" in morph.parse(line[i])[0].tag) and ("Name" in morph.parse(line[i+1])[0].tag)) or (("Name" in morph.parse(line[i+1])[0].tag) and ("Patr" in morph.parse(line[i+2])[0].tag)):
            if (i+3 == len(line)) or ("Surn" not in morph.parse(line[i+3])[0].tag and "Name" not in morph.parse(line[i+3])[0].tag):
                return True
            
    return False

#не все фамилии распознает как фамилии :( поэтому сделала двойное условие с or
#1335 строк, когда отфильтровала по наличию трех слов подряд, начинающихся с заглавной (это оценка сверху, поскольку лишние варианты типа Великая Отечественная Война тоже попали)
#979, когда отфильтровала по этому без OR, 1393 - иначе

check_for_single_name_in_line("Мемориальная доска Акулишнину Фёдору Васильевичу"), check_for_single_name_in_line("Мемориальная доска памяти ополченцев Народного комиссариата иностранных дел РФ")
```

```python
# pattern = r'\b[А-Я][а-я]+\s[А-Я][а-я]+\s[А-Я][а-я]+\b' #три слова подряд, начинающиеся с заглавной: ФИО
# filtered_df = df[df['Name'].str.contains(pattern)].copy()
filtered_df = df[df['Name'].apply(lambda x: check_for_single_name_in_line(x))]
filtered_df 
```

```python
def extract_single_name_from_line(line):
    line = line.split()
    for i in range(len(line)-2):
        # print(line[i], morph.parse(line[i])[0].tag)
        if (("Surn" in morph.parse(line[i])[0].tag) and ("Name" in morph.parse(line[i+1])[0].tag)) or (("Name" in morph.parse(line[i+1])[0].tag) and ("Patr" in morph.parse(line[i+2])[0].tag)):
            if (i+3 == len(line)) or ("Surn" not in morph.parse(line[i+3])[0].tag and "Name" not in morph.parse(line[i+3])[0].tag):
                return " ".join(line[i:i+3])
            
    return ""
```

```python
filtered_df['Name'] = filtered_df['Name'].apply(lambda x: extract_single_name_from_line(x)).copy()
filtered_df
```

```python
def get_gender(word):
    # word = word.split()[2] #отчество
    return morph.parse(word)[0].tag.gender
get_gender("Ивановой Ирине Николаевне")
```

```python
filtered_df['Gender'] = filtered_df['Name'].apply(lambda x: get_gender(x))
filtered_df
```

```python
def extract_coordinates(row):
    coordinates = row['geoData'].split('[')[1].split(']')[0].split(',')
    return pd.Series({'Latitude': float(coordinates[0]), 'Longitude': float(coordinates[1])})

# Apply the function to create new columns
filtered_df[['Latitude', 'Longitude']] = filtered_df.apply(extract_coordinates, axis=1)
filtered_df
```

Cоздаем табличку для того, чтобы сделать граф по странам (какие часто кооперируются друг с другом)

```python
mmkf = pd.read_csv("sysblok_movies_Moscow.csv")
mmkf['Countries'] = mmkf['Countries'].str.split(',')
mmkf
```

```python
countrypairs = pd.Series([])
countrypairs
```

```python
import itertools
# counter = 0
for index, row in mmkf.iterrows():
  countries = row['Countries']
  # counter += 1 - проверяла, что правда проходит по всему датасету
  # Generate all unique pairs of countries
  for country1, country2 in itertools.combinations(countries, 2):
      new_data = pd.Series([1], index=[tuple(sorted((country1, country2)))])
      countrypairs = pd.concat([countrypairs, new_data])
countrypairs
```

```python
countrypairs = countrypairs.reset_index()
countrypairs = countrypairs.groupby('index')[0].sum().sort_values(ascending=False).reset_index()
countrypairs
```

```python
countrypairs[['Country1', 'Country2']] = countrypairs['index'].apply(pd.Series)
countrypairs = countrypairs.drop(columns=['index'])
countrypairs
```

```python
countrypairs.to_csv('pairsofcountries.csv')
```
